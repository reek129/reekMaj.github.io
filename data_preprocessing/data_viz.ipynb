```python
import json

with open('dataviz/matches_Spain.json') as f:
    matches = json.load(f)
teams={}
with open('dataviz/teams.json') as f:
    team = json.load(f)
    for item in team:
        teams[str(item['wyId'])] = item
```

```python
sorted_matches=sorted(matches, key = lambda i: i['dateutc'])
#datetime.strptime(alert['alert_date'], "%Y-%m-%d %H:%M:%S")
'''
from datetime import datetime
for match in matches:
    match['date'] = datetime.strptime(' '.join(match['date'].split(' ')[0:3]),"%B %d, %Y")
'''
```

```python
#Find number of teams playing in the tournament
unique_teams={}
for match in matches:
    for team_id in list(match['teamsData'].keys()):
        unique_teams[str(team_id)]=1
    
```

```python
list(unique_teams.keys())
```

```python
import pandas as pd

matches_df=pd.DataFrame()
```

```python
win_points=3
draw_points=1
loose_points=0
```

```python
home_team=[]
home_goals=[]
away_team=[]
away_goals=[]
date2=[]
team=[]
week=[]
points=[]

for team_id in list(unique_teams.keys()):
    #print(team_id)
    team_points=[]
    cur_week=1
    for match in sorted_matches:
        teams_list=list(match['teamsData'].keys())
        
        #print(teams_list)
        #print(team_id)
        if team_id in teams_list:
            #print(team_id)
            #print(teams[teams_list[0]]['name'])
            
            if match['teamsData'][teams_list[0]]['side'] == 'home':
                home_team.append(str(teams[teams_list[0]]['name']))
                home_goals.append(match['teamsData'][teams_list[0]]['score'] + match['teamsData'][teams_list[0]]['scoreP'])
                away_team.append(teams[teams_list[1]]['name'].encode("utf-8").decode("utf-8"))
                away_goals.append(match['teamsData'][teams_list[1]]['score'] + match['teamsData'][teams_list[1]]['scoreP'])
            else:
                home_team.append(teams[teams_list[1]]['name'])
                home_goals.append(match['teamsData'][teams_list[1]]['score'] + match['teamsData'][teams_list[1]]['scoreP'])
                away_team.append(teams[teams_list[0]]['name'].encode("utf-8").decode("utf-8"))
                away_goals.append(match['teamsData'][teams_list[0]]['score'] + match['teamsData'][teams_list[0]]['scoreP'])
                
            date2.append(match['dateutc'])
            team.append(teams[team_id]['name'].encode("utf-8").decode("utf-8"))
            week.append(cur_week)
            cur_week=cur_week+1
            
            if len(team_points) == 0:
                if match['winner'] == int(team_id):
                    team_points.append(win_points)
                elif match['winner'] == 0:
                    team_points.append(draw_points)
                else:
                    team_points.append(loose_points)
            
            else:
                if match['winner'] == int(team_id):
                    team_points.append(team_points[-1]+win_points)
                elif match['winner'] == 0:
                    team_points.append(team_points[-1]+draw_points)
                else:
                    team_points.append(team_points[-1]+loose_points)
                
            
        else:
            continue
    
    points = points + team_points
    
```

```python
for cur_index, item in enumerate(team):
    if item == 'Atl\\u00e9tico Madrid':
        team[cur_index]='Atletico Madrid'
    if item == 'Deportivo La Coru\\u00f1a':
        team[cur_index]='Deportivo La Coruna'
    if item == 'Deportivo Alav\\u00e9s':
        team[cur_index]='Deportivo Alaves'
    if item == 'M\\u00e1laga':
        team[cur_index]='Malaga'
    if item == 'Legan\\u00e9s':
        team[cur_index]='Leganes'

for cur_index, item in enumerate(home_team):
    if item == 'Atl\\u00e9tico Madrid':
        home_team[cur_index]='Atletico Madrid'
    if item == 'Deportivo La Coru\\u00f1a':
        home_team[cur_index]='Deportivo La Coruna'
    if item == 'Deportivo Alav\\u00e9s':
        home_team[cur_index]='Deportivo Alaves'
    if item == 'M\\u00e1laga':
        home_team[cur_index]='Malaga'
    if item == 'Legan\\u00e9s':
        home_team[cur_index]='Leganes'
    
for cur_index, item in enumerate(away_team):
    if item == 'Atl\\u00e9tico Madrid':
        away_team[cur_index]='Atletico Madrid'
    if item == 'Deportivo La Coru\\u00f1a':
        away_team[cur_index]='Deportivo La Coruna'
    if item == 'Deportivo Alav\\u00e9s':
        away_team[cur_index]='Deportivo Alaves'
    if item == 'M\\u00e1laga':
        away_team[cur_index]='Malaga'
    if item == 'Legan\\u00e9s':
        away_team[cur_index]='Leganes'       
```

```python
matches_df['home_team']=home_team
matches_df['home_goals']=home_goals
matches_df['away_team']=away_team
matches_df['away_goals']=away_goals
matches_df['date2']=date2
matches_df['team']=team
matches_df['week']=week
matches_df['points']=points
```
